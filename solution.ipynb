```python
import pandas as pd
import numpy as np
import scipy as sp
```

```python
iris_df=pd.read_fwf('irisdata.txt', delimiter=',', names=['Sepal Length', 'Sepal Width', 'Petal length', 'Petal width', 'Class'])
iris_df.head()
```

```python
iris_df.replace('Iris-setosa', 'Iris-setosa, N = 50', inplace=True)
iris_df.replace('Iris-versicolor', 'Iris-versicolor, N = 50', inplace=True)
iris_df.replace('Iris-virginica', 'Iris-virginica, N = 50', inplace=True)
```

```python
summary_table = iris_df.groupby('Class').describe()
```

```python
summary_table = round(summary_table.transpose(), 2)
summary_table
```

```python
sepal_length = summary_table.loc[[('Sepal Length', 'count'), ('Sepal Length', 'mean'), ('Sepal Length', 'min'), ('Sepal Length', '50%'), ('Sepal Length', 'max'), ('Sepal Length', 'std')]]
sepal_length['Total, N = 150'] = sepal_length['Iris-setosa, N = 50'] + sepal_length['Iris-versicolor, N = 50'] + sepal_length['Iris-virginica, N = 50']
```

```python
sepal_length
```

```python
sepal_width = summary_table.loc[[('Sepal Width', 'count'), ('Sepal Width', 'mean'), ('Sepal Width', 'min'), ('Sepal Width', '50%'), ('Sepal Width', 'max'), ('Sepal Width', 'std')]]
sepal_width['Total, N = 150'] = sepal_width['Iris-setosa, N = 50'] + sepal_width['Iris-versicolor, N = 50'] + sepal_width['Iris-virginica, N = 50']
```

```python
sepal_width
```

```python
def sepal_length_perc_make(dataframe):
    less_5 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}
    between_5_6 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}
    between_6_7 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}
    more_7 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}

    for index, row in dataframe.iterrows():
        if row['Sepal Length'] < 5:
            less_5[row['Class']] += 1
        elif row['Sepal Length'] >= 5 and row['Sepal Length'] < 6:
            between_5_6[row['Class']] += 1
        elif row['Sepal Length'] >= 6 and row['Sepal Length'] < 7:
            between_6_7[row['Class']] += 1
        elif row['Sepal Length'] >= 7:
            more_7[row['Class']] += 1
    return [less_5, between_5_6, between_6_7, more_7]
```

```python
sepal_length_perc = pd.DataFrame(sepal_length_perc_make(iris_df), index = [('Sepal Length, %','< 5'), ('Sepal Length, %','>=5 AND < 6'), ('Sepal Length, %','>=6 AND < 7'), ('Sepal Length, %','>=7')])
sepal_length_perc['Total, N = 150'] = sepal_length_perc['Iris-setosa, N = 50'] + sepal_length_perc['Iris-versicolor, N = 50'] + sepal_length_perc['Iris-virginica, N = 50']
```

```python
sepal_length_perc
```

```python
sepal_length_perc[['Iris-setosa, %','Iris-versicolor, %', 'Iris-virginica, %', 'Total, %']] = round(sepal_length_perc[['Iris-setosa, N = 50', 'Iris-versicolor, N = 50', 
                                                                                                                       'Iris-virginica, N = 50', 'Total, N = 150']].apply(lambda x: (x/x.sum())*100, axis=0), 1)
sepal_length_perc
```

```python
sepal_length_perc['Iris-setosa, N = 50'] = sepal_length_perc['Iris-setosa, N = 50'].astype(str) +'(' + sepal_length_perc['Iris-setosa, %'].astype(str) + ')'
sepal_length_perc['Iris-versicolor, N = 50'] = sepal_length_perc['Iris-versicolor, N = 50'].astype(str) + '(' + sepal_length_perc['Iris-versicolor, %'].astype(str) + ')'
sepal_length_perc['Iris-virginica, N = 50'] = sepal_length_perc['Iris-virginica, N = 50'].astype(str) + '(' + sepal_length_perc['Iris-virginica, %'].astype(str) + ')'
sepal_length_perc['Total, N = 150'] = sepal_length_perc['Total, N = 150'].astype(str) + '(' + sepal_length_perc['Total, %'].astype(str) + ')'
```

```python
sepal_length_perc
```

```python
def sepal_width_perc_make(dataframe):
    less_3 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}
    between_3_35 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}
    between_35_4 = {'Iris-setosa, N = 50' : 0,
             'Iris-versicolor, N = 50' : 0,
             'Iris-virginica, N = 50': 0}
    
    for index, row in dataframe.iterrows():
        if row['Sepal Width'] < 3:
            less_3[row['Class']] += 1
        elif row['Sepal Width'] >= 3 and row['Sepal Width'] < 3.5:
            between_3_35[row['Class']] += 1
        elif row['Sepal Width'] >= 3.5 and row['Sepal Width'] < 4:
            between_35_4[row['Class']] += 1
            
    return [less_3, between_3_35, between_35_4]
```

```python
sepal_width_perc = pd.DataFrame(sepal_width_perc_make(iris_df), index = [('Sepal Width, %', '< 3'), ('Sepal Width, %', '>=3 AND < 3.5'), ('Sepal Width, %', '>=3.5 AND < 4')])
sepal_width_perc['Total, N = 150'] = sepal_width_perc['Iris-setosa, N = 50'] + sepal_width_perc['Iris-versicolor, N = 50'] + sepal_width_perc['Iris-virginica, N = 50']
```

```python
sepal_width_perc[['Iris-setosa, %','Iris-versicolor, %', 'Iris-virginica, %', 'Total, %']] = round(sepal_width_perc[['Iris-setosa, N = 50', 'Iris-versicolor, N = 50', 
                                                                                                                       'Iris-virginica, N = 50', 'Total, N = 150']].apply(lambda x: (x/x.sum())*100, axis=0), 1)
sepal_width_perc
```

```python
sepal_width_perc['Iris-setosa, N = 50'] = sepal_width_perc['Iris-setosa, N = 50'].astype(str) +'(' + sepal_width_perc['Iris-setosa, %'].astype(str) + ')'
sepal_width_perc['Iris-versicolor, N = 50'] = sepal_width_perc['Iris-versicolor, N = 50'].astype(str) + '(' + sepal_width_perc['Iris-versicolor, %'].astype(str) + ')'
sepal_width_perc['Iris-virginica, N = 50'] = sepal_width_perc['Iris-virginica, N = 50'].astype(str) + '(' + sepal_width_perc['Iris-virginica, %'].astype(str) + ')'
sepal_width_perc['Total, N = 150'] = sepal_width_perc['Total, N = 150'].astype(str) + '(' + sepal_width_perc['Total, %'].astype(str) + ')'
```

```python
summary_table_full = pd.concat([sepal_length, sepal_length_perc, sepal_width, sepal_width_perc], axis=0)
```

```python
summary_table_full = summary_table_full.drop(['Iris-setosa, %', 'Iris-versicolor, %', 'Iris-virginica, %', 'Total, %'], axis=1)
summary_table_full
```

```python
summary_table_full.loc[('Sepal Length', 'count')]['Iris-setosa, N = 50'] = '{0}'.format(50)
summary_table_full.loc[('Sepal Length', 'count')]['Iris-versicolor, N = 50'] = '{0}'.format(50)
summary_table_full.loc[('Sepal Length', 'count')]['Iris-virginica, N = 50'] = '{0}'.format(50)
summary_table_full.loc[('Sepal Length', 'count')]['Total, N = 150'] = '{0}'.format(150)

summary_table_full.loc[('Sepal Width', 'count')]['Iris-setosa, N = 50'] = '{0}'.format(50)
summary_table_full.loc[('Sepal Width', 'count')]['Iris-versicolor, N = 50'] = '{0}'.format(50)
summary_table_full.loc[('Sepal Width', 'count')]['Iris-virginica, N = 50'] = '{0}'.format(50)
summary_table_full.loc[('Sepal Width', 'count')]['Total, N = 150'] = '{0}'.format(150)
```
